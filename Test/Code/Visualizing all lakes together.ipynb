```r
library(tidyverse)
library(ggthemes)
theme_set(theme_few())
```

```r
vic <- read.csv("../Data/7_overall_summary/particulate_abs_melted_summary.csv", header=T)
fil <- read.csv("../Data/7_overall_summary/fil_abs_melted_summary.csv", header=T)
raw <- read.csv("../Data/7_overall_summary/raw_abs_melted_summary.csv", header=T)
```

```r
options(repr.plot.width=6, repr.plot.height=6, repr.plot.res = 300)
ggplot(raw, aes(x=wl, y=mean, color=lake)) +
    geom_line() +
    geom_errorbar(aes(ymin = mean+std, ymax = mean-std), width=2, color="black") +
    geom_point() +
    labs(y="Mean Raw Absorption [1/m]", x="Wavelength (nm)", color="Lake") +
    theme(legend.position="bottom")#+
    #facet_wrap(.~lake, ncol=2)
```

```r
options(repr.plot.width=6, repr.plot.height=6, repr.plot.res = 300)
ggplot(fil, aes(x=wl, y=mean, color=lake)) +
    geom_line() +
    geom_errorbar(aes(ymin = mean+std, ymax = mean-std), width=2, color="black") +
    geom_point() +
    labs(y="Mean Filtered Absorption [1/m]", x="Wavelength (nm)", color="Lake") +
    theme(legend.position="bottom")#+
    #facet_wrap(.~lake, ncol=2)
```
